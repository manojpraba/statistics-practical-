```python
sns.histplot(np.random.lognormal(3.0,1.0,1000),kde=True)
```

if log finction is applied it converted into normal distribution

```python
lognormal=np.random.lognormal(3.0,1.0,1000)
```

```python
sns.histplot(np.log(lognormal),kde=True)
```

QQ - plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
```

```python
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(np.log(lognormal))
```

```python
plot_data((lognormal))
```

Bernoli distribution
Outcomes are binary values (Yes/No) or (0/1)


```python
p=0.3
bernouli=np.random.choice([0,1],size=50,p=[1-p,p])
```

```python
bernouli
```

```python
sns.histplot(bernouli,bins=2,kde=True,color='pink')
```
